```python
import os, shutil
```

```python
path = r"C:\Users\jemca\Downloads\\"
```

```python
os.listdir(path)
```

```python
folder_names = ['CSV_Files', 'TEXT_Files', 'Images_Files', 'MP4_Files', 'PDF_Files', 'Zip_Files']

for folder in folder_names:
    if not os.path.exists(path + folder):
        os.makedirs(path + folder)
```

```python
file_names = os.listdir(path)
```

```python
for file in file_names:
    if '.csv' in file.lower() and not os.path.exists(path + 'CSV_Files\\' + file):
        shutil.move(path + file, path + 'CSV_Files\\' + file)
    elif '.png' in file.lower() and not os.path.exists(path + 'Images_Files\\' + file):
        shutil.move(path + file, path + 'Images_Files\\' + file)
    elif '.jpeg' in file.lower() and not os.path.exists(path + 'Images_Files\\' + file):
        shutil.move(path + file, path + 'Images_Files\\' + file)
    elif '.jpg' in file.lower() and not os.path.exists(path + 'Images_Files\\' + file):
        shutil.move(path + file, path + 'Images_Files\\' + file)
    elif '.txt' in file.lower() and not os.path.exists(path + 'TEXT_Files\\' + file):
        shutil.move(path + file, path + 'TEXT_Files\\' + file)
    elif '.mp4' in file.lower() and not os.path.exists(path + 'MP4_Files\\' + file):
        shutil.move(path + file, path + 'MP4_Files\\' + file)
    elif '.pdf' in file.lower() and not os.path.exists(path + 'PDF_Files\\' + file):
        shutil.move(path + file, path + 'PDF_Files\\' + file)
    elif '.zip' in file.lower() and not os.path.exists(path + 'Zip_Files\\' + file):
        shutil.move(path + file, path + 'Zip_Files\\' + file)
        
```
